```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#First graph - generating a range of number with np.linspace()
#applying cos(x)
x=np.linspace(1,10)
y=np.cos(x)
plt.plot(x,y)
plt.show()
```

```python
#plotting multiple graphs
z=np.sin(x)
plt.plot(x,y,x,z);plt.show()
```

```python
#creating label in the axes and graph title
plt.plot(x,y,x,z)
plt.xlabel('Time axis') # labelling x-axis
plt.ylabel('Magnitude') # labelling y-axis
plt.title('Sine and Cosine waves') # graph title
plt.show() # printing
```

```python
#To put legend, we can use the method plt.legend(['x','y'],loc=value)
#if change the loc number, change the position of the legend. loc=0 matplotlib choose
plt.plot(x,y,x,z)
plt.xlabel('Time axis')
plt.ylabel('Magnitude')
plt.title('Sine and Cosine waves')
plt.legend(['y','z'],loc=0)
plt.show()
```

```python
# we can plot the graphs besides each other using subplotting
 # plot 1
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Cosine plot')
# plot 2
plt.subplot(1,2,2)
plt.plot(x,z)
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Sine plot')
# avoid overallping plots
plt.tight_layout() 
plt.show()
```

```python
# plot 1
plt.subplot(1,2,1) 
plt.plot(x,y,'r-x') # red plot with -x marker
plt.title('Cosine plot')
# plot 2
plt.subplot(1,2,2) 
plt.plot(x,z,'g-o') # green plot with -o marker
plt.title('Sine plot')
plt.tight_layout()
```
